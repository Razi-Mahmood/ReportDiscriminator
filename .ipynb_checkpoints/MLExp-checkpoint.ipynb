```python
def print_k_map(hmap,k):
    i=0
    for key in hmap:
        if (i<k):
            print(i,key,hmap[key])
        i+=1
def print_k_set(hset,k):
    i=0
    for key in hset:
        if (i<k):
            print(i,key)
        i+=1

```
